```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Use this cell to change folder locations if needed

```python
data_file_location = 'data/processed_data.tsv'
demographic_destination = 'figures/'
demographics_file_name = 'distribution.png'
```

Run this cell to save the distribution plot

```python
# Load the data from the data file
data = pd.read_csv(data_file_location, sep='\t')

# Format the style of the figure
order = data['project_id'].value_counts().index.tolist()
sns.set_theme(style="whitegrid")
plt.figure(figsize=(16, 8))
sns.countplot(x="project_id", data=data, order=order)
plt.xlabel("Project ID")
plt.ylabel("Patients")
plt.title("Number of Patients in Each Cancer Type")
plt.xticks(rotation=45)

# Save and show the plot
plt.savefig(demographic_destination + demographics_file_name)
plt.show()
```
