```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Proteomics

```python
## read proteomics data
data = pd.read_csv('./data/proteomics_data.tsv', sep='\t')

## format
order = data['tumor_code'].value_counts().index.tolist()
sns.set_theme(style='whitegrid')
plt.figure(figsize=(16, 8))
sns.countplot(x='tumor_code', data=data, order=order)
plt.xlabel('Tumor Code')
plt.ylabel('Patients')
plt.xticks(rotation=45)
plt.tight_layout()

## save figure
plt.savefig('./figures/distribution_proteomics.png')
plt.show()
```
