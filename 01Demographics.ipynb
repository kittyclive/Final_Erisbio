```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Use this cell to change folder locations if needed

```python
data_file_location = 'Data/all_normalized_data.tsv'
demographic_destination = 'Figures/'
demographics_file_name = 'Demographics.jpg'
```

Run this cell to save the demographics data

```python
# Load the data from the data file
data = pd.read_csv(data_file_location, sep='\t')

# Format the style of the figure
sns.set(style="whitegrid")
plt.figure(figsize=(12, 5))
sns.histplot(data["project_id"], discrete=True, shrink=0.8)
plt.xlabel("Project ID")
plt.ylabel("Patients")
plt.title("Number of Patients in Each Cancer Type")
plt.xticks(rotation=45)

# Save and show the plot
plt.savefig(demographic_destination + demographics_file_name)
plt.show()
```
